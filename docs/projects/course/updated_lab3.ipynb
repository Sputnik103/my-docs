### TASK 1: Scrape Movie Reviews from DouBan

In this task, you will collect comments for two movies from DouBan.

#### Task Requirements:
- **Web Scraping Setup**: Configure the scraping setup using requests and BeautifulSoup.
- **Comment Extraction**: Extract the first 100 comments for each movie using the specified URLs.
- **Data Storage**: Store the comments in lists for further processing.

#### Implementation Tips:
- Avoid movies with too few comments, as they may not provide enough data for analysis.

#### Code Implementation:

```python
import requests
from bs4 import BeautifulSoup

def get_comments(movie_url, num_comments):
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(movie_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    comments = soup.find_all('div', {'class': 'comment'})[:num_comments]
    return [comment.find('p').text.strip() for comment in comments]

# TODO: 替换下面的电影序号为实际的豆瓣电影页面序号
movie1 = '36208094'
movie2 = '36369452'
#

movie1_comments = []
movie2_comments = []
num = 500
for i in range(0, num, 20):
    movie1_comments.extend(get_comments(f'https://movie.douban.com/subject/{movie1}/comments?start={i}&limit=20&status=P&sort=new_score', 20))
    movie2_comments.extend(get_comments(f'https://movie.douban.com/subject/{movie2}/comments?start={i}&limit=20&status=P&sort=new_score', 20))

print(len(movie1_comments))
print(len(movie2_comments))
print(movie1_comments[0])
print(movie2_comments[0])
```

```python
import jieba

# 定义一个函数用于文本的预处理
def preprocess_text(comments):
    # TODO: 加载停用词表
    with open('stopwords.txt', 'r', encoding='utf-8') as file:
        stopwords = set([line.strip() for line in file.readlines()])
        
    processed_comments = []
    for comment in comments:
        # TODO: 使用jieba.cut进行中文分词
        words = jieba.cut(comment, cut_all=False)
        #

        # TODO: 去除停用词和多余空格
        filtered = [word.strip() for word in words if word.strip() not in stopwords and word.strip() != '']
        #

        # TODO: 拼接分词结果为一个字符串，并添加到processed_comments中
        processed_comments.append(' '.join(filtered))
        #
        
    return processed_comments

# 对电影评论进行预处理
movie1_processed = preprocess_text(movie1_comments)
movie2_processed = preprocess_text(movie2_comments)

print(movie1_processed)
print(movie2_processed)
```

### TASK 3: Perform and Compare Topic Analysis

Utilize the Latent Dirichlet Allocation (LDA) model to analyze the themes of movie reviews. Compare the topics discovered for each movie to gain insights into the thematic differences.

#### Task Requirements:
- **Vectorization**: Convert the preprocessed text into a document-term matrix.
- **LDA Analysis**: Apply the LDA model to discover the underlying topics.
- **Result Interpretation**: Discuss the significance of the discovered topics.

#### Implementation Tips:
- Choose an appropriate number of topics to explore.
- Analyze the top words of each topic to interpret the themes accurately.

#### Code Implementation:

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# 定义一个函数来执行LDA主题模型分析
def lda_topic_modeling(processed_texts, num_topics):
    
    # TODO: 使用CountVectorizer对文本进行向量化
    vectorizer = CountVectorizer()
    doc_term_matrix = vectorizer.fit_transform(processed_texts)
    #

    # TODO: 使用LatentDirichletAllocation进行主题分析
    lda = LatentDirichletAllocation(n_components=num_topics, random_state=0)
    lda.fit(doc_term_matrix)
    #

    # 获取每个主题的词分布
    topic_words = {}
    words = vectorizer.get_feature_names_out()
    
    for topic, comp in enumerate(lda.components_):
        word_idx = comp.argsort()[-5:]  # 每个主题取权重最大的5个词
        topic_words[topic] = [words[i] for i in word_idx]
    return topic_words

# 对两部电影的评论进行主题分析
topics_movie1 = lda_topic_modeling(movie1_processed, 5)
topics_movie2 = lda_topic_modeling(movie2_processed, 5)

print("Movie 1 Topics:")
for topic, words in topics_movie1.items():
    print(f"Topic {topic}: {', '.join(words)}")

print("\nMovie 2 Topics:")
for topic, words in topics_movie2.items():
    print(f"Topic {topic}: {', '.join(words)}")
```

```python
# TODO: Transforming the document-term matrix to obtain topic distribution
doc_topic = lda.transform(dtm_tfidf)

# Checking the shape of the resulting distribution
print(doc_topic.shape)

# Displaying the dominant topic for the first 10 documents
for n in range(doc_topic[:10].shape[0]):
    dominant_topic = doc_topic[n].argmax()
    print(f"Document {n}: Dominant topic {dominant_topic}\n")
```
